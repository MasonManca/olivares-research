```python
import requests
import time
import matplotlib.pyplot as plt

# Set the number of requests to make and the base API URL
num_requests = 100
api_url = "https://api.openai.com/v1/engines/davinci-codex/completions"

# Set the prompt to use for the requests
prompt = "Hello, world!"

# Create empty lists to store the response times and request numbers
response_times = []
request_nums = []

# Make the requests and record the response times
for i in range(num_requests):
    start_time = time.time()
    response = requests.post(api_url, json={"prompt": prompt})
    end_time = time.time()
    response_time = end_time - start_time
    response_times.append(response_time)
    request_nums.append(i + 1)
    print(f"Request {i + 1}: {response_time} seconds")

# Plot the response times
plt.plot(request_nums, response_times)
plt.xlabel("Request Number")
plt.ylabel("Response Time (seconds)")
plt.title("ChatGPT AI Response Times")
plt.show()
```
